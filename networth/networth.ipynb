```python
import pandas as pd

pd.set_option('display.max_rows', None)
pd.set_option('display.max_columns', 1000)
pd.set_option('display.width', None)
pd.set_option('display.max_colwidth', None)
```

```python
df=pd.read_csv('blist.csv')
```

```python
df
```

```python
df['NetWorthNumber'] = df['NetWorth'].apply(lambda x: float(x[1:len(x)-1]) * 1000000000)
```

```python
def cannonic(name):
    return name.replace(' Jr', '').replace(' & family', '')


def get_last(name):
    return cannonic(name).rsplit(sep=' ', maxsplit=1)[1]


def get_first(name):
    return cannonic(name).rsplit(sep=' ', maxsplit=1)[0]
```

```python
df['FirstName'] = df['Name'].apply(get_first)
df['LastName'] = df['Name'].apply(get_last)
```

```python
df.head(500)
```

```python
from difflib import get_close_matches

def predict_net_worth(name):
    match = get_close_matches(name, df['LastName'], cutoff=0.1)
    return df[df['LastName'].isin(match)]['NetWorthNumber'].mean()
```
